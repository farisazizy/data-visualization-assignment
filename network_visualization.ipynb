```python
import networkx as nx
import matplotlib.pyplot as plt
```

Peta-Jawa-Barat.jpg

```python
# Buat Nodes
G = nx.Graph()
G.add_node("Sukabumi", pos=(12,16))
G.add_node("Bogor", pos=(11,20))
G.add_node("Cianjur", pos=(15,17))
G.add_node("Bekasi", pos=(13,25))
G.add_node("Karawang", pos=(17,25))
G.add_node("Purwakarta", pos=(20,21))
G.add_node("Bandung", pos=(23,16))
G.add_node("Subang", pos=(25,21))
G.add_node("Sumedang", pos=(27,17))

G.add_node("Garut", pos=(27,11))
G.add_node("Majalengka", pos=(31,17))
G.add_node("Tasikmalaya", pos=(31,10))
G.add_node("Ciamis", pos=(33,10))
G.add_node("Banjar", pos=(35,9))
G.add_node("Kuningan", pos=(34,15))
G.add_node("Cirebon", pos=(35,18))
G.add_node("Indramayu", pos=(32,24))

G.nodes()
pos = nx.get_node_attributes(G,'pos')
```

```python
plt.figure(figsize=(7, 5))
nx.draw(G, pos)
nx.draw_networkx_labels(G, pos, font_size=8)
plt.show()
```

```python
G.add_edge("Sukabumi", "Bogor")
G.add_edge("Sukabumi", "Cianjur")

G.add_edge("Cianjur", "Bogor")
G.add_edge("Cianjur", "Bandung")

G.add_edge("Bogor", "Bekasi")
G.add_edge("Bekasi", "Karawang")
G.add_edge("Purwakarta", "Karawang")
G.add_edge("Purwakarta", "Indramayu")

G.add_edge("Subang", "Majalengka")
G.add_edge("Subang", "Purwakarta")
G.add_edge("Subang", "Indramayu")

G.add_edge("Bandung", "Purwakarta")
G.add_edge("Bandung", "Sumedang")
G.add_edge("Bandung", "Subang")
G.add_edge("Bandung", "Garut")
G.add_edge("Bandung", "Tasikmalaya")

G.add_edge("Garut", "Tasikmalaya")
G.add_edge("Tasikmalaya", "Ciamis")
G.add_edge("Ciamis", "Banjar")
G.add_edge("Ciamis", "Kuningan")
G.add_edge("Majalengka", "Sumedang")
G.add_edge("Majalengka", "Kuningan")
G.add_edge("Majalengka", "Cirebon")
G.add_edge("Kuningan", "Cirebon")
G.add_edge("Indramayu", "Cirebon")

G.edges()
```

```python

pos_higher = {}
y_off = 0.8  # offset on the y axis

for k, v in pos.items():
    pos_higher[k] = (v[0], v[1] + y_off)

plt.figure(figsize=(7, 5))
nx.draw(G, pos)
nx.draw_networkx_labels(G, pos_higher, font_size=8)
plt.show()
```
